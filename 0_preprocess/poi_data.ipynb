```python
# Load the shapefile of LSOA of London
lsoa_gdf = gpd.read_file('../2_correlation/santander_service_area_theisson.geojson')

lsoa_gdf['SHAPE_area'] = lsoa_gdf.geometry.area
lsoa_gdf = lsoa_gdf[['inbound_counts', 'EndStation Id', 'name','SHAPE_area', 'geometry']]
```

## POI density calculation

```python


# Filter POIs into categories and store them in a dictionary
poi_gdfs = {
    'ShortAccommodation': ShortAccommodation,
    'catering': catering,
    'daily_shopping': daily_shopping,
    'nonessential_retail': nonessential_retail,
    'public_facilities': public_facilities,
    'sports': sports,
    'attractions': attractions,
    'indoor_entertain': indoor_entertain,
    'education': education,
    'offices': offices,
    'bus': bus,
    'metro': metro,
    'railway': railway
}


# Initialize an empty DataFrame to hold all densities
all_poi_densities = pd.DataFrame(index=lsoa_gdf['EndStation Id'].unique())
all_poi_densities.index.name = 'EndStation Id'

for poi_name, poi_gdf in poi_gdfs.items():
    if not poi_gdf.empty:  # Improved check for an empty GeoDataFrame
        # Perform spatial join to find LSOAs containing POIs
        joined_gdf = gpd.sjoin(lsoa_gdf, poi_gdf, how="inner", op='contains')
        poi_count = joined_gdf.groupby('EndStation Id').size().rename(f"{poi_name}_count")
        
        # Calculate density directly without adding count to lsoa_gdf
        # Need to ensure the 'SHAPE_area' column is handled correctly - assumed it's in 'lsoa_gdf'
        all_poi_densities[f"{poi_name}_count"] = poi_count
        all_poi_densities[f"{poi_name}_density"] = poi_count / lsoa_gdf.set_index('EndStation Id')['SHAPE_area']*1000000
       

# Ensure that 'EndStation Id' in 'lsoa_gdf' is not set as index to allow 'merge' operation
lsoa_gdf = lsoa_gdf.merge(all_poi_densities, on='EndStation Id', how='left').fillna(0)

lsoa_gdf.head(2)
```

```python
## Road density calculation
```

```python
roadnet = gpd.read_file('../GIS_file/road_network/roadnet_santander.shp') # polyline

# Calculate total road length within each LSOA unit
roadnet['length'] = roadnet.geometry.length

# Spatially join LSOA and roadnet data
joined_line = gpd.sjoin(lsoa_gdf, roadnet, how="inner", predicate="intersects")

# Sum the total road length within each LSOA unit
length_sum = joined_line.groupby('EndStation Id')['length'].sum().reset_index()

# Merge the sum result back to lsoa_gdf
lsoa_gdf = lsoa_gdf.merge(length_sum, on='EndStation Id', how='left')

# Set NaN values to 0 for LSOA units without matching roads
lsoa_gdf['length'].fillna(0, inplace=True)

# Calculate road density and adjust units to kilometers per square kilometer (km/km²)
lsoa_gdf['road_density'] = (lsoa_gdf['length'] / lsoa_gdf['SHAPE_area'])*1000

# Rename columns
#lsoa_gdf = lsoa_gdf.rename(columns={'length': 'road_length'})
lsoa_gdf = lsoa_gdf.drop(columns=['length'])
# View the results
lsoa_gdf.head(2)
```

```python
pop = gpd.read_file('../GIS_file/POI/OA_point.shp') # polyline

pop.head(1)
```

```python
import geopandas as gpd

# 假设lsoa_gdf和oa_points_gdf已经被正确加载
# 并且执行了必要的空间连接和USUALRES数值累加步骤

# 使用sjoin找出每个OA点落在哪个LSOA多边形内，并计算USUALRES之和
joined_gdf = gpd.sjoin(pop, lsoa_gdf, how='inner', op='within')
usual_res_sum = joined_gdf.groupby('EndStation Id')['USUALRES'].sum().reset_index()

# 将累加结果的列名从USUALRES更改为pop_count
usual_res_sum.rename(columns={'USUALRES': 'pop_count'}, inplace=True)

# 确保lsoa_gdf的'EndStation Id'字段类型与usual_res_sum的匹配
lsoa_gdf['EndStation Id'] = lsoa_gdf['EndStation Id'].astype(usual_res_sum['EndStation Id'].dtype)

# 合并usual_res_sum回lsoa_gdf，基于'EndStation Id'
lsoa_gdf = lsoa_gdf.merge(usual_res_sum, on='EndStation Id', how='left')

# 如果有LSOA中没有任何OA点，pop_count的和可能是NaN，可以选择将这些NaN值替换为0
lsoa_gdf['pop_count'] = lsoa_gdf['pop_count'].fillna(0)

# 计算人口密度，并将结果存储在新列'pop_density'中
lsoa_gdf['pop_density'] = lsoa_gdf['pop_count'] / lsoa_gdf['SHAPE_area'] * 1000000

# 最后，lsoa_gdf包含了一个新列'pop_count'，表示落入每个LSOA多边形内的OA点中USUALRES的数值之和
lsoa_gdf.head(2)
```

## calculate POI entrophy

In this code, the calculate_poi_entropy function calculates the POI mixing entropy for the incoming row (representing a geographic unit). It first finds all columns ending in _count that contain counts of different POI types. It then calculates the proportion of these POI types in the total and uses these proportions to calculate the mixing entropy. Finally, we apply this function to each row of lsoa_gdf, calculate the POI mixing entropy for each geographic unit, and save the results in a new column poi_entropy.
<br><br>** If the total POI count of a geographic unit is 0, the function will return 0 as its mixing entropy value, since such a geographic unit has practically no POI diversity to speak of. In addition, to avoid errors due to zeros when calculating logarithms, we add a very small number **np.finfo(float).eps** to the ratio.<br><br>


Shannon's Diversity Index (H') is a measure of diversity within a dataset that accounts for both abundance and evenness of the categories present. In the context of Geographic Information Systems (GIS), it can be used to quantify the diversity and balance of Point of Interest (POI) types within a geographical unit. A higher Shannon index indicates a more diverse distribution, where POI types are more evenly spread, while a lower index indicates a less diverse distribution, possibly dominated by one or a few types.<br><br>
The formula for Shannon's Diversity Index (H'):

**\[ H' = -\sum_{i=1}^{S} p_i \ln(p_i) \]**

Where:
- \(H'\) is the Shannon Diversity Index,
- \(S\) is the total number of categories within the system (for example, the total number of different POI types),
- \(p_i\) is the relative proportion of the \(i\)th category within the system (i.e., the number of entities in that category divided by the total number of entities),
- \(\ln\) represents the natural logarithm.


```python
import numpy as np
import geopandas as gpd

# 定义相关POI计数列
poi_columns = [
    'ShortAccommodation_count', 'catering_count', 'daily_shopping_count',
    'nonessential_retail_count', 'public_facilities_count', 'sports_count',
    'indoor_entertain_count', 'education_count', 'offices_count',
    'bus_count', 'metro_count', 'railway_count'
]

# 选择这些POI列，并确保它们是numpy数组
counts = lsoa_gdf[poi_columns].values

# 计算总数以避免除以零，使用keepdims=True保持维度以便于广播
total_counts = np.sum(counts, axis=1, keepdims=True)

# 计算每个POI类型的比例，使用where条件避免除以零的错误
proportions = np.divide(counts, total_counts, where=total_counts!=0)

# 为了安全进行对数计算，使用np.maximum避免零值
proportions_safe = np.maximum(proportions, np.finfo(float).eps)

# 计算混合熵
entropy = -np.sum(proportions_safe * np.log(proportions_safe), axis=1)

# 将计算结果存储在新的列'poi_entropy'中
lsoa_gdf['poi_entropy'] = entropy

# 展示计算结果
print(lsoa_gdf[['poi_entropy']].head())
```

## landuse ratio in each station service area

```python
import geopandas as gpd

# 假设 lsoa_gdf 原本应该是一个GeoDataFrame，但可能在处理过程中丢失了地理信息
# 确保 'geometry' 列仍然存在于 lsoa_gdf 中
# 如果 'geometry' 列不存在，你需要先解决这个问题

# 将 lsoa_gdf 转换回 GeoDataFrame
if not isinstance(lsoa_gdf, gpd.GeoDataFrame):
    lsoa_gdf = gpd.GeoDataFrame(lsoa_gdf, geometry='geometry')

# 现在你可以安全地继续使用 lsoa_gdf 进行空间连接了
# 例如：
# joined_gdf = gpd.sjoin(gdf, lsoa_gdf, how="inner", predicate="intersects")
```

```python
import warnings

# Suppress warnings from the Fiona and GeoPandas libraries
warnings.filterwarnings('ignore', category=UserWarning, module='fiona')
warnings.filterwarnings('ignore', category=UserWarning, module='geopandas')


# Streamline area calculation for each GeoDataFrame
categories_gdfs = {
    'commercial': gpd.read_file('../GIS_file/POI/landuse_commercial1.shp'),
    'residential': gpd.read_file('../GIS_file/POI/landuse_resident.shp'),
    'leisurepark': gpd.read_file('../GIS_file/POI/leisure_park.shp'),
    'retail': gpd.read_file('../GIS_file/POI/landuse_retail.shp')
}

for name, gdf in categories_gdfs.items():
    gdf['area'] = gdf.geometry.area

# Define the categories with the updated GeoDataFrames having 'area' calculated
categories = {
    'commercial': categories_gdfs['commercial'],
    'residential': categories_gdfs['residential'],
    'leisurepark': categories_gdfs['leisurepark'],
    'retail': categories_gdfs['retail']
}

def process_land_use_category(gdf, lsoa_gdf, category_name):
    # Spatial join
    joined_gdf = gpd.sjoin(gdf, lsoa_gdf, how="inner", predicate="intersects")
    
    # Sum areas within each LSOA unitimport warnings
    area_sum = joined_gdf.groupby('EndStation Id')['area'].sum().reset_index(name=f'{category_name}_area')
    
    # Merge the sum back into lsoa_gdf and fill NaN values with 0
    lsoa_gdf = lsoa_gdf.merge(area_sum, on='EndStation Id', how='left').fillna({f'{category_name}_area': 0})
    
    return lsoa_gdf

# Assuming lsoa_gdf is your target GeoDataFrame and it's already defined
for category_name, gdf in categories.items():
    lsoa_gdf = process_land_use_category(gdf, lsoa_gdf, category_name)

# Corrected print statement to display the head of lsoa_gdf
lsoa_gdf.head(2)
```

```python
# 计算人口密度，并将结果存储在新列'pop_density'中
lsoa_gdf['commercial_ratio'] = lsoa_gdf['commercial_area'] / lsoa_gdf['SHAPE_area']
lsoa_gdf['residential_ratio'] = lsoa_gdf['residential_area'] / lsoa_gdf['SHAPE_area']
lsoa_gdf['leisurepark_ratio'] = lsoa_gdf['leisurepark_area'] / lsoa_gdf['SHAPE_area']
lsoa_gdf['retail_ratio'] = lsoa_gdf['retail_area'] / lsoa_gdf['SHAPE_area']
lsoa_gdf.describe()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from scipy.stats import norm

# Assuming lsoa_gdf is your DataFrame and you have imported required libraries

columns_to_visualize = ['inbound_counts']

for column in columns_to_visualize:
    plt.figure(figsize=(10, 6))
    sns.histplot(lsoa_gdf[column], kde=True, color='blue', label='Histogram', stat='density')
    plt.title(column.replace('_', ' ').capitalize())
    plt.xlabel(column)
    plt.ylabel('Density')
    
    # Fit a normal distribution to the data
    mu, std = norm.fit(lsoa_gdf[column])
    xmin, xmax = plt.xlim()
    x = np.linspace(xmin, xmax, 100)
    p = norm.pdf(x, mu, std)
    
    plt.plot(x, p, 'k', linewidth=2, label='Normal Distribution')
    plt.legend()
    plt.show()
```
